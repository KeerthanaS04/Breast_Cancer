```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
new_df=df.groupby(['Age']).count()
```

```python
new_df.head()
```

```python
plt.figure(figsize=(30,50))
sns.set_theme(style="ticks",font_scale=1.2)
plt.subplots_adjust(wspace=0.4,hspace=0.2)
plt.subplot(2,1,1)
plt.xticks(rotation=45)
plt.plot(new_df['Survival Months'],label='Survival Months',color='red',linestyle='dashed',linewidth=3,marker='*')
plt.subplot(2,1,2)
plt.xticks(rotation=45)
plt.plot(new_df['6th Stage'],label='6th Stage',color='black',linestyle='dashed',linewidth=3,marker='o')
plt.show()
```
